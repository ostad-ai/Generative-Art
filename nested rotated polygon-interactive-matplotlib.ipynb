```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib notebook
fig, ax = plt.subplots(figsize=(7.5,5))
plt.axis([0,1,0,1])
plt.title('Interactive nested rotated polygons')
scale=100.
nGon=5
def polygon(x,y,n=5,radius=.1,angle=0.):
    teta=np.pi/n
    corners=np.zeros((n,2))
    angle=.2*np.pi*np.sin(1.*angle)
    for i in range(n):
        corners[i,:]=radius*np.array([np.cos(angle+2*i*teta),
                                    np.sin(angle+2*i*teta)])
    corners[:,0]+=x; corners[:,1]+=y
    return corners
def onclick(event):
    angle=0.; 
    for r in range(10,1,-1):
        corners=polygon(event.xdata,event.ydata,nGon,r/scale,angle)
        color='#'+''.join(np.random.choice(list('0123456789abcdef'),6))
        pat=plt.Polygon(corners,color=color)
        ax.add_patch(pat)
        angle+=.05

cid = fig.canvas.mpl_connect('button_press_event', onclick)
```
