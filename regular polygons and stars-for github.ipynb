```python
from math import sin,cos,pi,gcd
from matplotlib import pyplot as plt
```

```python
# regular polygon with n_gon corners
def polygon_regular(n_gon=3,radius=10.):
    vertices=[]
    d_angle=2.*pi/n_gon
    angle=0.
    for i in range(n_gon):
        x,y=radius*cos(angle), radius*sin(angle)
        vertices.append([x,y])
        angle+=d_angle
    return vertices

# regular star with n_gon corners and m_gon steps
def star_regular(n_gon=3,m_gon=2,radius=10.):
    vertices=polygon_regular(n_gon,radius)
    vs_n_m=[]
    i_start=0
    for m in range(m_gon):
        for i in range(i_start,n_gon,m_gon):
            vs_n_m.append(vertices[i])
        i_start=(i+m_gon)%n_gon
    return vs_n_m
```

```python
max_num=12 # max number of vertices
i=0; plt.figure(figsize=(20,15))
for p in range(3,max_num):
    for q in range(1,int(p/2)+1):
        if gcd(p,q)==1 or q==1:
            i+=1
            plt.subplot(4,5,i)
            xs,ys=zip(*star_regular(p,q))
            plt.fill(xs,ys,fc='aqua',ec='k')
            if q==1:
                plt.title(f'Regular polygon ({p})')
            else:
                plt.title(f'Regular star ({p},{q})')
plt.show()
```
