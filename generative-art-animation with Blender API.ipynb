```python
# uncomment and run the following code in the Text Editor of Blender
import bpy
bpy.ops.mesh.primitive_cube_add()
ao=bpy.context.active_object #active object
modifier_sub=ao.modifiers.new('My Subdivide 1','SUBSURF')
modifier_sub.levels=3
bpy.ops.object.shade_smooth()
modifier_disp=ao.modifiers.new('My Displacment 1','DISPLACE')
texture=bpy.data.textures.new('My Texture 1','DISTORTED_NOISE')
texture.noise_scale=1
modifier_disp.texture=texture
mat=bpy.data.materials.new(name='My Material 1')
mat.diffuse_color=(1.,127/255,80/255,1.) #coral
ao.data.materials.append(mat)
key_frames=[0,50,100,150,200,250]
scales=[1,2,4,8,16,1]
for sc,fr in zip(scales,key_frames):
    modifier_disp.texture.noise_scale=sc
    modifier_disp.texture.keyframe_insert(data_path='noise_scale',frame=fr)
```
