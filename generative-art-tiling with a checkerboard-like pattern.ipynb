### Generative Art: Tiling with a checkerboard-like pattern

```python
from matplotlib import pyplot as plt
import numpy as np
from pycolors import css_colors #pycolors is in the github
```

```python
width,height=800,200
check_img=np.zeros((height,width,3),dtype='int')
sx,sy=30,30 
sx2,sy2=sx//2,sy//2
sx4,sy4=sx//4,sy//4
for x in range(width):
    for y in range(height):
        if ((x//sx)+(y//sy))%2==0:
            check_img[y,x,:]=css_colors['crimson']
            if (((x+sx4)//sx2)+((y+sx4)//sx2))%2==0:
                check_img[y,x,:]=css_colors['antiquewhite']
for x in range(width):
    for y in range(height):
        if not ((x//sx)+(y//sy))%2==0 and \
        (((x+sx4)//sx2)+((y+sx4)//sx2))%2==0:
            check_img[y,x,:]=css_colors['darkcyan']
plt.figure(figsize=(16,4))
plt.imshow(check_img)
plt.axis('off')
# Uncomment below to save the image in your folder
#plt.imsave('checker_color.png',check_img.astype('uint8'),dpi=150)
plt.show()
```
