```python
import random
from matplotlib import pyplot as plt
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x; self.y=y
        
def chaosGame(Ts,Probs,iter=10000,
    palette=['red','blue','green','gold','cyan','magenta'],discard=20):
    P=Point(0,0)
    points=[]
    for i in range(iter):
        index=random.choices(range(len(Ts)),Probs)[0]
        T=Ts[index]
        P.x,P.y=T[0]*P.x+T[1]*P.y+T[4],T[2]*P.x+T[3]*P.y+T[5]
        points.append((P.x,P.y,palette[index]))
    return points[discard:]
```

```python
Ts=[[0.,0.,0.,.16,0.,0.],
   [.2,-.26,.23,.22,0.,1.6],
   [-.15,.28,.26,.24,0.,.44],
   [.75,.04,-.04,.85,0.,1.6]]
Probs=[.1,.08,.08,.74]
palette=['green','green','green','green']
points=chaosGame(Ts,Probs,100000,palette)
xs,ys,colors=zip(*points)
plt.scatter(xs,ys,c=colors,s=.1,alpha=.8)
plt.axis('off')
#plt.savefig('./chaosgame-myleaf.png',
#            bbox_inches='tight', pad_inches=0,dpi=150)
plt.show()   
```
