```python
import numpy as np; from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def disk(cZ=1.,diskR=1.,cX=0.,cY=0.):
    R=np.linspace(0,diskR,nPoints); Beta=np.linspace(0,2*np.pi,nPoints)
    R,Beta=np.meshgrid(R,Beta,indexing='ij')
    X,Y=cX+R*np.cos(Beta),cY+R*np.sin(Beta)
    Z=cZ*np.ones_like(X); return X,Y,Z
def cylinder(zBegin=0.,zEnd=1.,radius=5.,cX=0.,cY=0.,nPoints=50):
    Teta=np.linspace(0.,2.*np.pi,nPoints);
    Z=np.linspace(zBegin,zEnd,nPoints)
    Z,Teta=np.meshgrid(Z,Teta,indexing='ij')
    X=cX+radius*np.cos(Teta); Y=cY+radius*np.sin(Teta)
    return X,Y,Z
```

```python
nPoints=50;nCyn=10; dZ=1.;zStart=0.; rStart=1.
colors=['yellow','red','cyan']; nColors=len(colors)
fig=plt.figure(figsize=(12,7)); ax=fig.gca(projection='3d')
for i in range(nCyn):
    zBegin=zStart+i*dZ; zEnd=zBegin+dZ
    radius=rStart/float(i+1); X,Y,Z=disk(zBegin,radius)
    ax.plot_surface(Z,Y,X,color=colors[i%nColors],alpha=.5)
    X,Y,Z=cylinder(zBegin,zEnd,radius)
    ax.plot_surface(Z,Y,X,color=colors[i%nColors],alpha=1.)
    X,Y,Z=disk(zEnd,radius);
    ax.plot_surface(Z,Y,X,color=colors[i%nColors],alpha=.5)
ax.view_init(15,45)
plt.title(f'Gabriel\'s Wedding Cake with {nCyn} cylinders')
plt.show()
```
