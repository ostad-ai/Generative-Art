# Truchet tiling with diagonal lines for creating labyrinths
We can use two kinds of diagonal lines for creating **labyrinths**. The type of diagonal is chosen randomly for each position.
<br> In fact, we use tiles with a square shape having one of its diagonal drawn. Thus, we are tiling the plane wth squares of the same size.

```python
# importing required modules
import pygame as pg
import random
from matplotlib import pyplot as plt
# you should have the numpy module installed
```

```python
# the class for tile definition 
class Tile:
    def __init__(self,x=0,y=0,size=20):
        self.x=x; self.y=y
        self.size=size; 
        orient=random.randint(0,1)
        self.colorW=[255,255,255]
        self.colorB=[0,50,100]
        self.image=pg.Surface((size,size))
        self.image.fill(self.colorW)
        if orient==0:
            pg.draw.line(self.image,self.colorB,(0,0),(size,size),width=4)
        else:
            pg.draw.line(self.image,self.colorB,(size,0),(0,size),width=4)
    def draw(self,surface):
        surface.blit(self.image,(self.x,self.y))
        
def generateTiles(rows=25,cols=25,tile_size=30):
    pg.init()
    tiles=[Tile(col*TileSize,row*TileSize,TileSize) for col in range(cols) for row in range(rows)]
    screen=pg.display.set_mode((WIDTH,HEIGHT),pg.HIDDEN)
    for tile in tiles:
        tile.draw(screen)
    image=pg.surfarray.array3d(screen).swapaxes(0,1)
    pg.quit()
    return image
```

```python
# calling the functions for creating tiles for the truchet-based patterns
# each call to the function generateTiles builds a different labyrinth
Rows,Cols,TileSize=20,20,50
WIDTH,HEIGHT=Cols*TileSize,Rows*TileSize
image=generateTiles(Rows,Cols,TileSize)
plt.figure(figsize=(12,7))
plt.subplot(121);plt.imshow(image);plt.axis('off')
image=generateTiles(Rows,Cols,TileSize)
plt.subplot(122);plt.imshow(image);plt.axis('off')
plt.show()
```
