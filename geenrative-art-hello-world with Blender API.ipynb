```python
# uncomment and run the following code in the Text editor of Blender
import bpy
from random import randint,random
back=bpy.data.worlds["World"].node_tree.nodes["Background"]
back.inputs[0].default_value = (.05,.05,.05, 1)
bpy.ops.mesh.primitive_plane_add()
bpy.ops.transform.resize(value=(50,50,1))
mat=bpy.data.materials.new(name="material_bg")
mat.diffuse_color=(1.,215/255,0,1) #gold
bpy.context.active_object.data.materials.append(mat)
number=300
for i in range(number):
    x=randint(-15,15); y=randint(-15,15)
    z=randint(0,15);  r=random()
    g=random(); b=random()
    mat=bpy.data.materials.new(name="material"+str(i))
    mat.diffuse_color=(r,g,b,1)
    bpy.ops.mesh.primitive_monkey_add(location=(x,y,z))
    bpy.context.active_object.data.materials.append(mat)
```
