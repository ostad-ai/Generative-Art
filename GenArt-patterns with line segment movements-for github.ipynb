# Generative Art:
### Pattern generation with line segment movements

<br>Here, we use line segments to create symmetric patterns. The endpoints of a line segment move over two different circles and the traces they leave create the pattern.

```python
# importing required libraries
from math import pi, sin, cos
from matplotlib import pyplot as plt
```

```python
# The function that defines two circles over which endpoints move
# and the frequency of the movements
# you can play with the parameters to get different patterns
def patternGeneration(radius1=73,freq1=11,radius2=379,freq2=1,dt=.005,tmax=1):
        lines=[]
        t=0 #time
        while t<tmax:
            x1,y1=radius1*cos(2*pi*freq1*t),radius1*sin(2*pi*freq1*t)
            x2,y2=radius2*cos(2*pi*freq2*t),radius2*sin(2*pi*freq2*t)
            lines.append([(x1,x2),(y1,y2)])
            t+=dt
        return lines
linSegs=patternGeneration()
for line in linSegs:
    plt.plot(line[0],line[1],c='k')
plt.axis('off'); plt.show()
```
